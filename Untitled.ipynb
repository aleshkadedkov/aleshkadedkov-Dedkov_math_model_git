```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры проекта
initial_investment = 100000  # Первоначальные инвестиции
cash_flows = np.array([30000, 40000, 50000, 60000, 70000])  # Денежные потоки за 5 лет
discount_rate = 0.1  # Ставка дисконтирования (10%)

# Функция для расчета NPV
def npv(rate, cash_flows, initial_investment):
    total_value = np.sum(cash_flows / (1 + rate) ** np.arange(1, len(cash_flows) + 1))
    return total_value - initial_investment

# Построение графика NPV в зависимости от ставки дисконтирования
rates = np.linspace(0, 0.5, 100)  # Ставки от 0% до 50%
npv_values = [npv(rate, cash_flows, initial_investment) for rate in rates]

plt.figure(figsize=(10, 6))
plt.plot(rates, npv_values, label='NPV', color='blue')
plt.axhline(0, color='red', linestyle='--', label='NPV = 0')
plt.title('График NPV в зависимости от ставки дисконтирования')
plt.xlabel('Ставка дисконтирования')
plt.ylabel('NPV')
plt.legend()
plt.grid()
plt.show()
```
